```python
%matplotlib inline
import matplotlib.pyplot as plt
from skimage import io, data, feature, color
import numpy as np

lane1 = io.imread('images/lane_detection/f00000.png')
lane2 = io.imread('images/lane_detection/f00050.png')
lane3 = io.imread('images/lane_detection/f00090.png')

eye1 = io.imread('images/eye_tracking/0000.jpeg')
eye2 = io.imread('images/eye_tracking/0050.jpeg')
eye3 = io.imread('images/eye_tracking/0090.jpeg')

fig = plt.figure(figsize=(15, 10))
ax11 = plt.subplot(2, 3, 1)
ax12 = plt.subplot(2, 3, 2)
ax13 = plt.subplot(2, 3, 3)
ax21 = plt.subplot(2, 3, 4)
ax22 = plt.subplot(2, 3, 5)
ax23 = plt.subplot(2, 3, 6)



ax11.imshow(lane1)
ax12.imshow(lane2)
ax13.imshow(lane3)

ax21.imshow(eye1)
ax22.imshow(eye2)
ax23.imshow(eye3)
```

```python
### my solution
def detect_edges(image):
    lane1_gray = color.rgb2gray(image)
    return feature.canny(lane1_gray, sigma=0.8)

edge_image = detect_edges(lane1)
# io.imshow(detect_edges(lane2))

def hough_transform(image):
    alpha_axis = np.linspace(0,np.pi, 180)
    unit_vectors = np.column_stack((np.cos(alpha_axis), np.sin(alpha_axis)))
    # cos_values = np.cos(alpha_axis)
    # sin_values = np.sin(alpha_axis)
    rho_max = np.sqrt(image.shape[0]**2 + image.shape[1]**2)
    rho_axis = np.linspace(-rho_max, rho_max, int(2*rho_max))
    acc_array = np.zeros((len(rho_axis), len(alpha_axis)))
    for y_idx, row in enumerate(image):
        for x_idx, _ in enumerate(row):
            if image[y_idx,x_idx]:
                for alpha_idx, unit_vector in enumerate(unit_vectors):
                    rho = np.dot(unit_vector, np.array([x_idx, y_idx]))
                    rho_index = np.argmin(np.abs(rho_axis - rho))
                    acc_array[rho_index, alpha_idx] += 1
    return acc_array, rho_axis, alpha_axis

acc, rho_table, alpha_table = hough_transform(edge_image)
max_index_flat = np.argmax(acc)
rho_index, alpha_index = np.unravel_index(max_index_flat, acc.shape)
print(rho_index, alpha_index)
```

```python
# dieser Code wurde als Musterlsung von Sebastian Oltmanns zur Verfgung gestellt und von Tim renoviert

%matplotlib inline
from skimage.draw import line
from skimage import io
import math
import numpy as np

def draw_line_hessian_normal(image, a, r):

    (dimy,dimx) = image.shape

    #r = x*cosa + y*sina

    # compute start and end point of line
    x0 = 0
    y0 = round((r - x0*math.cos(math.radians(a)))/math.sin(math.radians(a)))
    x1 = dimx
    y1 = round((r - x1*math.cos(math.radians(a)))/math.sin(math.radians(a)))

    liney,linex = line(y0,x0,y1,x1)

    ret = np.copy(image)
    for yy in range(0,liney.size-1):
        if (liney[yy] > 0) and (liney[yy] < dimy-1):
            ret[liney[yy],linex[yy]] = 1
    return ret

image = np.zeros((100,100))
image = draw_line_hessian_normal(image, 20, 60)
image = draw_line_hessian_normal(image, 20, 10)
image = draw_line_hessian_normal(image, 110, 0)
image = draw_line_hessian_normal(image, -70, 10)
io.imshow(image)
```
